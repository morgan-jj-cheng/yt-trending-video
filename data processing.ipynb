```python
# data from 0901 - 0228
start_date = datetime(2024, 9, 1)
end_date = datetime(2025, 2, 28)

# save dfs into their own category
df_dict = {
    "gaming":[], 
    "movies":[],
    "music":[]
}

files = ["gaming", "movies", "music"]
for name in files:
    for date in pd.date_range(start_date, end_date):
        link = f"pages/data/{name}/{name}_{date.strftime('%Y%m%d')}.csv"
        if os.path.exists(link):
            try:
                df_single = pd.read_csv(link)
                df_dict[name].append(df_single)
            except Exception as e:
                print(f"Error reading {link}: {e}")
        else:
            print(f"File not found: {link}")

            
#print(df_dict)

# combine all df
df_gaming_0228 = pd.concat(df_dict["gaming"], ignore_index=True)
df_movies_0228 = pd.concat(df_dict["movies"], ignore_index=True)
df_music_0228 = pd.concat(df_dict["music"], ignore_index=True)
```

```python
# combine 0831 and 0228 data
df_gaming_0831 = pd.read_csv("pages/data/gaming_0831.csv")
df_movies_0831 = pd.read_csv("pages/data/movies_0831.csv", encoding="cp1252")
df_music_0831 = pd.read_csv("pages/data/music_0831.csv", encoding="cp1252")


df_gaming = pd.concat([df_gaming_0831, df_gaming_0228], ignore_index=True)
df_movies = pd.concat([df_movies_0831, df_movies_0228], ignore_index=True)
df_music = pd.concat([df_music_0831, df_music_0228], ignore_index=True)
```

## Cleaning
### Gaming Dataframe

```python
df_gaming.sort_values("publishedDate", ascending=True).head()
```

```python
# structure
print(df_gaming.columns)
print(df_gaming.shape)
```

```python
# information
print(df_gaming.info())
print(df_gaming.describe())
```

```python
# missing value
print(df_gaming.isna().sum())
```

```python
# duplicate value
df_gaming.duplicated(subset="title").sum()
```

```python
df_gaming = df_gaming.groupby("title").agg(
    description=("description", "first"), 
    publishedDate=("publishedDate", "min"), 
    channelName	=("channelName", "first"), 
    views=("views", "max"), 
    duration=("duration", "first"), 
    isShort=("isShort", "first")
).reset_index()
```

```python
# check final length   
print(len(df_gaming))  #8987 - 1022
```
